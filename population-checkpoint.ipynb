#### Continent with the highest average population growth rate between 1990 and 2020?

```python
# Calculate the average growth rate for each continent
data['avg_growth_rate'] = ((data['2020_pop'] / data['1990_pop']) ** (1/30)) - 1

# Group by continent and calculate the mean of average growth rates
continent_avg_growth = data.groupby('Continent')['avg_growth_rate'].mean()

# Find the continent with the highest average growth rate
continent_highest_growth = continent_avg_growth.idxmax()
highest_growth_rate = continent_avg_growth.max()

print(f"The continent with the highest average growth rate between 1990 and 2020 is {continent_highest_growth} with an average growth rate of {highest_growth_rate:.2%}")


```

#### Countries that exibit outlier-like behavior in terms of their population growth rates?

```python
# Create a box plot
plt.figure(figsize=(10, 6))
plt.boxplot(data['growth_rate'], vert=False)
plt.xlabel('growth_rate')
plt.title('Population Growth Rate Distribution')
plt.tight_layout()

# Annotate outlier values
outliers = data[data['growth_rate'] > data['growth_rate'].quantile(0.75) + 1.5 * (data['growth_rate'].quantile(0.75) - data['growth_rate'].quantile(0.25))]

for index, row in outliers.iterrows():
    plt.annotate(row['country'], xy=(row['growth_rate'], 1), xytext=(10, 10), textcoords='offset points', arrowprops=dict(arrowstyle='->', color='red'))

plt.show()
```

#### Population growth rates of African countries compared to Asian countries

```python

# Separate African and Asian countries
african_df = data[data['Continent'] == 'Africa']
asian_df = data[data['Continent'] == 'Asia']

# Group and aggregate data
african_growth = african_df.groupby('Continent')['growth_rate'].aggregate('sum')
asian_growth = asian_df.groupby('Continent')['growth_rate'].aggregate('sum')

# Create a comparison plot
plt.figure(figsize=(15,8))
fig, ax = plt.subplots()
ax.bar(african_growth.index, african_growth, label='African Countries', color='blue')
ax.bar(asian_growth.index, asian_growth, label='Asian Countries', color='green')
ax.set_title('Comparison of Growth Rates: African vs. Asian Countries')
ax.set_xlabel('Continent')
ax.set_ylabel('Mean Growth Rate')
ax.legend()
plt.show()

asian_growth,african_growth 


```

#### Visualisation of global population 

```python
global_pop_percentage = data.groupby('Continent')['world_pop_percentage'].aggregate('sum').sort_values(ascending = False)
plt.figure(figsize = (15,8))
plt.bar(global_pop_percentage.index,global_pop_percentage.values, color = 'orange', width = 0.5)
plt.xlabel('Continents')
plt.ylabel('Total Continental population percentage')
plt.gca().invert_xaxis()
plt.show()

```

#### Countries that exhibited outlier-like behavior in their population growth between 1980 and 2020

```python
# Calculate population growth rate
data['PopulationGrowthRate'] = (data['2020_pop'] - data['1990_pop'])

# Create a violin plot
plt.figure(figsize=(7, 12))
sns.violinplot(data=data, y='PopulationGrowthRate',split = True)
plt.title('Population Growth Rate Distribution (1990-2020)')
plt.ylabel('Population Growth Rate (%)')
# plt.xscale('log')
# plt.yscale('log')

# Annotate top two outlier countries
outliers = data.nlargest(2, 'PopulationGrowthRate')

for index, row in outliers.iterrows():
    plt.annotate(row['country'], xy=(0, row['PopulationGrowthRate']), xytext=(20, 5), textcoords='offset points',arrowprops=dict(facecolor='black', shrink=0.05), fontsize=10, color='red', rotation=45)

plt.show()
```

#### continental relationship between a country’s population growth and population density.

```python
# Group by continent and calculate mean density and growth rate
continental_groups= data.groupby('Continent').agg({
    'density': 'mean',
    'growth_rate': 'mean'
}).reset_index()

# Create scatter plot to visualize the relationship
plt.figure(figsize=(10, 6))
plt.scatter(continental_groups['density'], continental_groups['growth_rate'], color='blue')
plt.title('Population Growth vs. Population Density by Continent')
plt.xlabel('Population Density')
plt.ylabel('Population Growth Rate')
plt.grid(True)
plt.show()

```

#### Correlation between a country’s land area and its population growth in the last three decades.

```python
correlation_data = data[["area", "growth_rate"]].corr()

# Plot correlation matrix using a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_data, annot=True, cmap="coolwarm", fmt=".2f", center=0)
plt.title("Correlation between Land Area and Population Growth Rate")
plt.show()
```
